Having tried with rolling window of 10 we found that the we can smoothen the noise in a very much better way compared to the rolling window size of 3 or 5. But the problem here is that if we start increasing the rolling window just to make the data smooth, we might not be able to reflect the recent changes in our predictions fast. So we should look at a method that smoothens the noise by giving more importance to the recent values compared to the past values which can be done using exponential smoothing

```python
# Writing a function to define the exponential smoothing

def exponential_smoothing(data, alpha):
    smoothed = [data[0]]  # Initialize S0 with the first data point
    for t in range(1, len(data)):
        smoothed_value = alpha * data[t] + (1 - alpha) * smoothed[t-1]
        smoothed.append(smoothed_value)
    return smoothed
```

```python
# Visualising the exponential smoothing curve for first row

plt.plot(exponential_smoothing(row,0.5))
```

In the above exponential smoothing we have given equal importance to both the historical data and also the current data

Here if we compare Moving Average 5 with the exponential smoothing we can say that Moving Average has smoothened the Noise in a better way than the exponential smoothing with giving equal importance to the historical data and the current data, So we can try the Moving Average 5 for our prediction

```python
# Transpose the data to treat columns as rows for easier rolling
data_t = df_withnoise.T

# Perform rolling average column-wise
window_size = 5
for i in range(window_size, len(data_t)):
    # Update the current column based on the mean of the previous 5 columns
    data_t.iloc[i] = data_t.iloc[i-window_size:i].mean()

# Transpose back to original format
smoothed_data = data_t.T
```
